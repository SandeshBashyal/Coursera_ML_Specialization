```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
myfile='heart.csv'
heart_data=pd.read_csv(myfile)
```

```python
heart_data.head()
```

```python
heart_data.tail()
```

Here are the visualization codes

```python
plt.figure(figsize=(16,6))
plt.title("Heart Attack Analysis and Prediction Data Set")
sns.lineplot(data=heart_data)
```

```python
list(heart_data.columns)
```

```python
sns.lineplot(data=heart_data['age'], label="Age")
sns.lineplot(data=heart_data['sex'], label="Sex")
sns.lineplot(data=heart_data['output'], label="Output")
```

```python
plt.figure(figsize=(10,6))
plt.title("Heart Attack Analysis and Prediction Data Set")
sns.barplot(x=heart_data.sex, y=heart_data['cp'])
plt.ylabel("CP")
plt.xlabel("SEX")
```

```python
plt.figure(figsize=(14,7))
plt.title("Heart Attack Analysis and Prediction Data Set")
sns.heatmap(data=heart_data, annot=True)
plt.xlabel("Features of Analysis")
```

```python
sns.scatterplot(x=heart_data['chol'], y=heart_data['thalachh'])
```

```python
sns.regplot(x=heart_data['chol'], y=heart_data['thalachh'])
```

```python
sns.scatterplot(x=heart_data['chol'], y=heart_data['thalachh'],hue=heart_data['sex'])
```

```python
sns.lmplot(x="chol", y="thalachh", hue="sex", data=heart_data)
```

```python
sns.swarmplot(x=heart_data['sex'],
              y=heart_data['thalachh'])
```

```python
sns.histplot(heart_data['oldpeak'])
```

```python
sns.kdeplot(data=heart_data['oldpeak'],fill=True)
```

```python
sns.jointplot(x=heart_data['oldpeak'], y=heart_data['restecg'], kind="kde")
```

```python
sns.histplot(data=heart_data, x='oldpeak', hue='sex')
plt.title("Histogram of Oldpeak, by sex")
```

```python
sns.kdeplot(data=heart_data, x='oldpeak', hue='sex', fill=True)
plt.title("Histogram of Oldpeak, by sex")
```
